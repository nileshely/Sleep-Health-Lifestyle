```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("/kaggle/input/sleep-health-and-lifestyle-dataset/Sleep_health_and_lifestyle_dataset.csv")
```

```python
plt.figure(figsize=(8, 6))
df['Gender'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='lightblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Sleep Duration'], bins=15, color='lightgreen')
plt.title('Sleep Duration Distribution')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Quality of Sleep', data=df, color='lightcoral')
plt.title('Quality of Sleep Distribution')
plt.xlabel('Quality of Sleep (scale: 1-10)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Physical Activity Level'], bins=15, color='lightyellow')
plt.title('Physical Activity Level Distribution')
plt.xlabel('Physical Activity Level (minutes/day)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Stress Level'], shade=True, color='lightgrey')
plt.title('Stress Level Distribution')
plt.xlabel('Stress Level (scale: 1-10)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='BMI Category', data=df, palette='pastel')
plt.title('BMI Category Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()
```

```python
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)

plt.figure(figsize=(10, 6))
sns.violinplot(x='Systolic_BP', data=df, palette='muted')
plt.title('Systolic Blood Pressure Distribution')
plt.xlabel('Systolic Blood Pressure')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Heart Rate'], bins=15, color='lightpink')
plt.title('Heart Rate Distribution')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Frequency')
plt.show()
```
